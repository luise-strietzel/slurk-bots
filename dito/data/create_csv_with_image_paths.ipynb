# creating csv files with the image paths

This notebook is used to fill csv files with urls to image pairs. They will be used to display images with (or without) differences to a pair of players in the dito game. 

```python
#imports
import csv
from PIL import Image
import requests
from os import listdir
from os.path import isfile, join
```

The images were manually added to the github-repository. There are two main folder <a href='https://github.com/luise-strietzel/images_temp/tree/main/form1'> **form1**</a> and <a href="https://github.com/luise-strietzel/images_temp/tree/main/form2">**form2**</a>. The folders contain images selected according to the two different task formulations. The images in **form1** are suitable for the task formulation "*Identify the difference together"*, whereas the image scenes in **form2** are rather suitable for the formulation "*Decide together whether or not there is a difference in your views!"*. 

```python
#the path both folders share
#basic_path = 'https://raw.githubusercontent.com/luise-strietzel/images_temp/main/form'
basic_path = 'https://raw.githubusercontent.com/luise-strietzel/slurk-bots/development/image_selection/form'
```

```python
#obtain a list of all the file names for formulation 1: 
files_form1 = [f for f in listdir('D:/image_selection/form1/source/')]

#and for form2:
files_form2 = [f for f in listdir('D:/image_selection/form2/source/')]
#print the length of the list and a few list entries to see if it worked
print(len(files_form2))
print(files_form2[1:5])
```

```python
#bringing together the conrete file names with the path segments leading to the github url of the images:
form1_list=[]
for i in files_form1: 
    source_ = '1/source/'
    target_ = '1/target/'
    
    entire_path_source = basic_path+source_+i
    entire_path_target = basic_path+target_+i
    form1_list.append([entire_path_source,entire_path_target])
```

```python
#show some lines and c+p the path into a browser tap to see if worked correctly
form1_list[1:5]
```

```python
#bringing together the conrete file names with the path segments leading to the github url of the images:
form2_list=[]
for i in files_form2: 
    source_ = '2/source/'
    target_ = '2/target/'
    
    entire_path_source = basic_path+source_+i
    entire_path_target = basic_path+target_+i
    form2_list.append([entire_path_source,entire_path_target])
```

```python
form2_list[1:5]
```

```python
#writing the entries for form1 into a csv file:
with open('form1_git.csv','w', newline='') as file:
    writer = csv.writer(file, delimiter=',')
    writer.writerows(form1_list)
```

```python
#and the entries for from2
with open('form2_git.csv','w', newline='') as file:
    writer = csv.writer(file, delimiter=',')
    writer.writerows(form2_list)
```
